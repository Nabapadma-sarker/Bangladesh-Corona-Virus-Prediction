```python
#gradient Descent
def gradientDecent(X, y, theta, alpha, iterations):

    m = len(y)
    bestJ = CostFunction(X, y, theta);
    for iter in range(1, iterations, 1):
        

        # ====================== YOUR CODE HERE ======================
        # Instructions: Perform a single gradient step on the parameter vector
        #               theta. 
        #
        # Hint: While debugging, it can be useful to print out the values
        #       of the cost function (computeCost) and gradient here.
        #
        #np.dot(Theta,X)
        #theta = theta - alpha*((((theta'*X')'.-y)'*X)'/m)        
        z = 0
        for i in range(0, m, 1):
                 z = z + (np.dot(theta,X[i])-y[i])*X[i]
        z = (alpha*z)/m
        theta = theta - z
        J = CostFunction(X, y, theta)
        if bestJ > J:
            bestJ = J
        

    return (theta, J)


#Cost function
def CostFunction(X, y, theta):
        m = len(y) # number of training examples

        # You need to return the following variables correctly 
        J = 0

        # ====================== YOUR CODE HERE ======================
        # Instructions: Compute the cost of a particular choice of theta
        #               You should set J to the cost.

        #J = sum(((theta'*X')'-y).^ 2)/(2*m)
                 
        for i in range(0, m, 1):
                 J = J + (np.dot(theta,X[i])-y[i])** 2
        J = J/(2*m)
        
        return J



# =========================================================================
```

```python
y = data['new deaths']
y1 = data['new cases']
X = np.arange(len(y))
theta = np.zeros(2)
all_ones = np.ones(len(y))
m = len(y)

mul_X = np.vstack((all_ones,X))
tran_X = mul_X.T;

iterations = 5000
alpha = 0.0000000000000001
  
theta, J = gradientDecent(tran_X, y, theta, alpha, iterations)

        
print(J)
print(theta)
```

```python
Y =[]
for j in range(1,m,1):
    Y.append(np.dot(theta,tran_X[j]))
Y.insert(0,0)
print(X)
print(Y)
```

```python
plt.figure(figsize=(25,10))
# plotting a bar chart 
plt.title('New Death Graph')
plt.xlabel('Date')
plt.ylabel('new deaths')
#plt.plot(color='red')
Y1 = np.multiply(1000000000,Y)


# plotting the actual points as scatter plot 
plt.scatter(X, y, color = "m", marker = "o", s = 30) 

# plotting the regression line 
plt.plot(X, Y1, color = "g") 

plt.plot(X, y, color='red')
plt.savefig('./newLearningDeathsLinegraph.png')


# function to show plot 
plt.show()
```

```python
# Fitting Linear Regression to the dataset 
from sklearn.linear_model import LinearRegression 
# Fitting Polynomial Regression to the dataset 
from sklearn.preprocessing import PolynomialFeatures 
  
poly = PolynomialFeatures(degree = 4) 
X_poly = poly.fit_transform(tran_X) 
#print(X_poly[5]) 
#print(tran_X.shape)
#poly.fit(X_poly, y) 

lin2 = LinearRegression() 
lin2.fit(X_poly, y) 

plt.figure(figsize=(25,10))
# Visualising the Polynomial Regression results 
plt.scatter(X, y, color = 'blue')
  
plt.plot(X, lin2.predict(X_poly), color = 'red')
plt.plot(X, y, color='black')
plt.title('Polynomial Regression')
plt.xlabel('Date')
plt.ylabel('New Deaths')
  
plt.show()
```
